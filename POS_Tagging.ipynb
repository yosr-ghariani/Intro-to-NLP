```python
#importation
import nltk
from nltk import *
from nltk.corpus import brown
nltk.download('brown')
nltk.download('universal_tagset')
```

```python
#reading the tagged words of brown
brown_tagged =brown.tagged_words(tagset='universal')
print(brown_tagged)
```

```python
len(brown_tagged)
```

```python
#parcours des verbs dans brown
verb=set()
for i in range(len(brown_tagged)):
    if brown_tagged[i][1]=='VERB':
      verb.add(brown_tagged[i][0])
```

```python
verb
```

```python
#parcours des noms dans brown
noun=[]
for i in range(len(brown_tagged)):
    if brown_tagged[i][1]=='NOUN':
      noun.append(brown_tagged[i][0])
print(noun)
```

```python
#calcul des occurences des noms et extraire les 10 les plus frequents
from collections import Counter
noun_freq = Counter(noun)
most_common_nouns = noun_freq.most_common(10)
print(most_common_nouns)
```

```python
#unigram, bigram and trigram taggers (bigram models)
from nltk.tag import UnigramTagger, BigramTagger, TrigramTagger

train_sents = []
test_sents = []
# Split the data: 90% for training, 10% for testing
train_size = int(len(brown_tagged) * 0.9)
# Utiliser une boucle for pour diviser les données
for i in range(len(brown_tagged)):
    if i < train_size:
        train_sents.append(brown_tagged[i])  # Ajouter à l'ensemble d'entraînement
    else:
        test_sents.append(brown_tagged[i])   # Ajouter à l'ensemble de test

# Train a Unigram and Bigram model
unigram_tagger = UnigramTagger(train_sents)
bigram_tagger = BigramTagger(train_sents, backoff=unigram_tagger)
trigram_tagger = TrigramTagger(train_sents, backoff=bigram_tagger)

# Check how accurate the Bigram model is on the test data
unigram_accuracy = unigram_tagger.evaluate(test_sents)
print(f"unigram Tagger Accuracy: {unigram_accuracy:.4f}")
bigram_accuracy = bigram_tagger.evaluate(test_sents)
print(f"Bigram Tagger Accuracy: {bigram_accuracy:.4f}")
trigram_accuracy = trigram_tagger.evaluate(test_sents)
print(f"Trigram Tagger Accuracy: {trigram_accuracy:.4f}")
```
